```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import yfinance as yf
```

```python
start = '2015-01-01'
end = '2025-06-01'

stock = 'IBM'

data = yf.download(stock, start, end)
```

```python
data.reset_index(inplace=True)
```

```python
data
```

```python
ma_100_days = data.Close.rolling(100).mean()
ma_200_days = data.Close.rolling(200).mean()

```

```python
plt.figure(figsize=(12,6))
plt.plot(ma_100_days, 'g')
plt.plot(data.Close, 'b')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(ma_100_days, 'g')
plt.plot(ma_200_days, 'r')
plt.plot(data.Close, 'b')
plt.show()
```

```python
data.dropna(inplace=True)
```

```python
data_train = pd.DataFrame(data.Close[0: int(len(data)*0.80)])
data_test = pd.DataFrame(data.Close[int(len(data)*0.80): len(data)])
```

```python
data_train.shape[0]
```

```python
data_test.shape[0]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
scaled_data_train = scaler.fit_transform(data_train)
scaled_data_test = scaler.fit_transform(data_test)
```

```python
import numpy as np
x_train = []
y_train = []
for i in range(100, len(data_train)):
    x_train.append(scaled_data_train[i-100: i])
    y_train.append(scaled_data_train[i, 0])


x_train = np.array(x_train)
y_train = np.array(y_train)
```

```python
from keras.layers import Dense, Dropout, LSTM
from keras.models import Sequential
```

```python
model = Sequential()
model.add(LSTM(units=50, activation ='relu', return_sequences=True, input_shape=((x_train.shape[1], 1))))
model.add(Dropout(0.2))
model.add(LSTM(units=60, activation = 'relu', return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(units=80, activation = 'relu', return_sequences=True))
model.add(Dropout(0.4))
model.add(LSTM(units=120, activation = 'relu'))
model.add(Dropout(0.5))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(x_train, y_train, epochs=50)
```

```python
model.summary()
```

```python
past_100_days = scaled_data_train[-100:]
```

```python
scaled_data_test = pd.concat([data_train, data_test], axis=0)
```

```python
data_test = scaler.fit_transform(data_test)
```

```python
import numpy as np
x_train = []
y_train = []
for i in range(100, len(data_test)):
    x_train.append(data_test[i-100: i])
    y_train.append(data_test[i, 0])

x_train = np.array(x_train)
y_train = np.array(y_train)
```

```python
y_predict =model.predict(x_train)
```

```python
scale = 1/scaler.scale_
```

```python
y_predict = y_predict * scale
```

```python
y_train = y_train*scale
```

```python
plt.figure(figsize=(10,6))
plt.plot(y_predict, 'g', label = 'Predicted Price')
plt.plot(y_train, 'b', label = 'Original Price')
plt.xlabel("Time")
plt.ylabel("Price")
plt.legend()
plt.show()
```
